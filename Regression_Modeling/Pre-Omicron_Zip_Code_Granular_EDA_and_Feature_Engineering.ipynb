```python
import pandas as pd

df = pd.read_csv('data/merged_zip_data.csv')
df.head()
```

```python
zipcodes_served = df.groupby('WRRF Name')['Zipcode'].unique().reset_index()
zipcodes_served
```

```python
zipcodes_served['Zip Codes Served'] = zipcodes_served['Zipcode'].apply(lambda x: ', '.join(map(str, x)))
zipcodes_served = zipcodes_served.drop(columns=['Zipcode'])


merged_df = df.merge(zipcodes_served, on='WRRF Name', how='left')


merged_df.drop('Zipcode', axis=1, inplace=True)

merged_df.drop_duplicates(subset='WRRF Name').reset_index(drop=True)
```

```python
# merged_df.to_csv('data/facility_zip_data.csv', index=False)
```

```python
# pre_omi_df = pd.read_csv('data/pre_omi')
```

```python
def split_and_clean(zipcodes):
    if isinstance(zipcodes, str):
        return [int(zipcode.strip()) for zipcode in zipcodes.split(',')]
    return []

merged_df['Zip_Codes_Served'] = merged_df['Zip Codes Served'].apply(split_and_clean)
zips_exploded_df = merged_df.explode('Zip_Codes_Served')
zips_exploded_df = zips_exploded_df.rename(columns={'Zip_Codes_Served': 'Individual Zip Code'})
zips_exploded_df.drop('Zip Codes Served', axis=1, inplace=True)
zips_exploded_df.head()
```

```python
zips_exploded_df.to_csv('data/exploded_zips.csv', index=False)
```
